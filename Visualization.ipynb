```python
dims = [256,512,1028]
times_CPU = [0.05409, 0.048089, 0.0558209 ]
times_CUDA = [0.0035072,0.00365056, 0.00365024]
times_shared = [0.0029344, 0.00369536, 0.00285376]
times_excess = [0.00902954, 0.0091873,0.00955904]


plt.plot(dims, times_CPU, label='CPU')
plt.plot(dims, times_CUDA, label='Regular CUDA')
plt.plot(dims, times_shared, label='Shared mem CUDA')
plt.plot(dims, times_excess, label='Excess copy CUDA')
plt.xlabel('Block size')
plt.ylabel('Time (ms)')
plt.xscale('log') 
plt.yscale('log')
plt.legend()
plt.grid(True) 
plt.savefig('Speedups.png')
plt.show()
```
